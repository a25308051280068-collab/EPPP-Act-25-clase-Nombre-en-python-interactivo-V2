```python
class Deporte:
    # El método __init__ define los atributos
    def __init__(self):
        self.nombre = ""
        self.tipo = ""
        self.jugador = ""

    # Función para leer datos (Interacción)
    def leer_datos(self):
        print(f"\n--- Ingresando datos del deporte ---")
        self.nombre = input("Nombre del deporte: ")
        self.tipo = input("Tipo (Individual/Equipo): ")
        self.jugador = input("Cantidad de jugadores: ")
        print(">> Mensaje: ¡Datos guardados exitosamente!")

    # Función para mostrar datos (Resultado)
    def mostrar_datos(self):
        print(f"\n--- Ficha Técnica ---")
        print(f"Deporte: {self.nombre}")
        print(f"Categoría: {self.tipo}")
        print(f"Participantes: {self.jugador}")
        print(f">> Mensaje: Mostrando información en pantalla...")

# --- CREACIÓN DE OBJETOS (Instanciación) ---

# Creamos el objeto deporte1
deporte1 = Deporte()
deporte1.leer_datos() # Pedirá datos al usuario

# Creamos el objeto deporte2
deporte2 = Deporte()
deporte2.leer_datos() # Pedirá datos al usuario

# --- RESULTADOS ---
print("\n" + "="*30)
print("REPORTE FINAL DE OBJETOS")
print("="*30)

deporte1.mostrar_datos()
deporte2.mostrar_datos()

print("programa terminado")
print("programa realizado por Christian Garcia NL = 21")
```
